## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
# Choose the column
col = "windspeed"

# Q1 and Q3
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)

# IQR
IQR = Q3 - Q1

# Bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

print(f"üîç Outlier Detection in '{col}' column:")
print(f"Q1: {Q1:.4f}, Q3: {Q3:.4f}, IQR: {IQR:.4f}")
print(f"Lower Bound: {lower_bound:.4f}, Upper Bound: {upper_bound:.4f}")
print(f"Number of outliers: {outliers.shape[0]}")
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
# Step 1: Map numeric season codes to names
season_map = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
df["season_name"] = df["season"].map(season_map)

# Step 2: Group by season and sum total rentals (cnt)
season_totals = df.groupby("season_name")["cnt"].sum()

# Step 3: Find the season with the highest rentals
most_rented_season = season_totals.idxmax()
max_rentals = season_totals.max()

# Step 4: Print the result
print("üìä Total Rentals by Season:")
print(season_totals)
print(f"\n‚úÖ The season with the highest rentals is: **{most_rented_season}** with {max_rentals} total rentals.")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
# Group by hour and sum total rentals
hourly_totals = df.groupby("hr")["cnt"].sum()

# Find the hour with maximum rentals
peak_hour = hourly_totals.idxmax()
max_rentals = hourly_totals.max()

# Display result
print("üìä Total Rentals by Hour of the Day:")
print(hourly_totals)

print(f"\n‚úÖ The peak hour with the highest bike rentals is: {peak_hour}:00 with {max_rentals} total rentals.")
```

## Q.18 In which day of a week bikes rents out most?

```python
# Map weekday numbers to names
weekday_map = {
    0: "Sunday", 1: "Monday", 2: "Tuesday", 3: "Wednesday",
    4: "Thursday", 5: "Friday", 6: "Saturday"
}
df["weekday_name"] = df["weekday"].map(weekday_map)

# Group by weekday and sum rentals
weekday_totals = df.groupby("weekday_name")["cnt"].sum()

# Sort and find peak day
peak_day = weekday_totals.idxmax()
max_rentals = weekday_totals.max()

# Display result
print("üìÖ Total Rentals by Day of Week:")
print(weekday_totals)

print(f"\n‚úÖ The day with the highest rentals is: **{peak_day}** with {max_rentals} total rentals.")
```

## Q.19 In which hour Casual users rents bikes the most?

```python
# Group by hour and sum casual rentals
casual_hourly = df.groupby("hr")["casual"].sum()

# Find the hour with the maximum casual rentals
peak_hour = casual_hourly.idxmax()
max_casual_rentals = casual_hourly.max()

# Display result
print("üö≤ Casual User Rentals by Hour:")
print(casual_hourly)

print(f"\n‚úÖ Casual users rent bikes the most at: {peak_hour}:00 with {max_casual_rentals} total rentals.")
```
