```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r'C:\Users\anupa\Downloads\car_age_price.csv')
```

```python
df.isnull().sum()
```

```python
x = df[['Year']]  # Independent variable
y = df['Price']   # Dependent variable
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
intercept = model.intercept_
slope = model.coef_[0]
print(f"Price = {intercept:.2f} + {slope:.2f} * Year")
```

```python
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, model.predict(x), color='red', label='Linear Fit')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Best Linear Fit: Year vs Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
predicted_price_2022 = model.predict([[2022]])
print("Predicted Price for 2022 model:", predicted_price_2022[0])
```
